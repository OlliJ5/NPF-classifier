```python
import pandas as pd
import numpy as np

df = pd.read_csv('npf_train.csv')
```

```python
# Luodaan sarake class2 jossa on nonevent kun class4 == nonevent ja event kun class4 != nonevent

class2 = np.array([1]*df.shape[0])
class2[df['class4']=='nonevent'] = 0

df['class2'] = class2

df.head()
```
